### 3D-Fit of Spectra with respect to Air Mass, Height, and Wavlength

```python
#fitting multiple Air Mass and heights to AM, height, and wavelength
low_wlindex = 20
high_wlindex = 200
airmasses = [1.0,1.25,1.5,1.75,2.0,3.0]
height_numbers = [6,8,10,12,14,16,18,20]
irrvheight_1_0 = np.loadtxt(open("am1p0_irr.csv"), delimiter=",", skiprows=1)
irrvheight_1_25 = np.loadtxt(open("am1p25_irr.csv"), delimiter=",", skiprows=1)
irrvheight_1_5 = np.loadtxt(open("am1p5_irr.csv"), delimiter=",", skiprows=1)
irrvheight_1_75 = np.loadtxt(open("am1p75_irr.csv"), delimiter=",", skiprows=1)
irrvheight_2_0 = np.loadtxt(open("am2p0_irr.csv"), delimiter=",", skiprows=1)
irrvheight_3_0 = np.loadtxt(open("am3p0_irr.csv"), delimiter=",", skiprows=1)
wavelength = irrvheight_1_0[low_wlindex:high_wlindex,0].T
#to do direct instead, should be able to change indices in following line to "4::2"
allglobals = np.concatenate((irrvheight_1_0[low_wlindex:high_wlindex,3::2], irrvheight_1_25[low_wlindex:high_wlindex,3::2],
                            irrvheight_1_5[low_wlindex:high_wlindex,3::2],irrvheight_1_75[low_wlindex:high_wlindex,3::2],
                            irrvheight_2_0[low_wlindex:high_wlindex,3::2],irrvheight_3_0[low_wlindex:high_wlindex,3::2])
                            , axis = 1).T
allglobals = array(allglobals)
```

```python
N_wls = len(wavelength)

wls = array([wavelength]*len(height_numbers)*len(airmasses))
heights = []
for h in height_numbers:
    heights = heights+[h]*N_wls
heights = array([heights]*len(airmasses))
airmasslist = []
hwls = len(height_numbers)*N_wls
for a in airmasses:
    airmasslist = airmasslist+([a]*hwls)
airmasslist = array(airmasslist)
allglobals = array(allglobals)
```

```python
allgs_photons_pernm = allglobals* wls/1e9 / units("h*c").value.to_base_units().magnitude
allgs_photons_cum = np.cumsum(allgs_photons_pernm, axis=1)*10


#plot(wls.T,allgs_photons_cum.T)
```

```python
am_h_spec_cs, am_h_spec_err = gpfit(log([wls.flatten(),heights.flatten(),airmasslist]), 
                                    -log(allgs_photons_cum.flatten()),4,"SMA")
print am_h_spec_err

                                    
```

### Junctions and Cells and Efficiency Oh My

```python
class CumulativeSpectrum(Model):
    "TODO: Take arguments like latitude, elevation"
    def setup(self):
        self.phi = Variable("\\Phi_G", "count/s/m^2", "Solar irradiance (photons)")
        self.lam = Variable("\\lambda", "nm", "Maximum wavelength of light accumulated")
        
        phi_nd = self.phi/units("count/s/m^2")
        lam_nd = self.lam/units("nm")
        
        return [phi_nd**-0.011146 >= (0.592622*lam_nd**-0.00417149
                                      + 1.38955e+80*lam_nd**-32.8796
                                      + 131556*lam_nd**-2.6098)]
        

class Junction(Model):
    "TODO: add materials, fill factor, absorption"
    def setup(self):
        self.lam_g = Variable("\\lambda_g", "nm", "Band gap wavelength")
        self.E_gV = Variable("E_{gV}", "eV", "Band gap energy")
        self.I = Variable("I_{sc}", "A/m^2", "Short-circuit current")
        self.V = Variable("V_{oc}", "V", "Open-circuit voltage")
        self.phi = Variable("\\Phi_G", "count/s/m^2", "Incident photons")
        self.G = Variable("G", "count/s/m^2", "Generation rate")
        self.W_oc = Variable("W_{oc}", 0.3, "V", "Band-gap offset")
        self.A = Variable("A", 1, "-", "Optical absorption")
        self.QE = Variable("QE", 1, "-", "Quantum efficiency")
        
        return [self.E_gV == 1240.*units("eV nm")/self.lam_g,  # TODO: where does 1240 come from?
                self.V + self.W_oc <= self.E_gV/units("e"),
                self.G == self.phi * self.A,
                self.I == self.QE*self.G*units("e"),
               ]

# TODO: fix printing bug with sub-named constraintsets

class Cell(Model):
    "TODO: add QE model"
    def setup(self, N_junctions):
        self.eta = Variable("\\eta", "-", "Cell efficiency")
        self.I = Variable("I_{sc}", "A/m^2", "Short-circuit current")
        self.V = Variable("V_{oc}", "V", "Open-circuit voltage")

        total_irrad = sum(np.diff(wavelength) * globaltilt[:-1])*units("W/m^2")
        
        with Vectorize(N_junctions):
            self.jns = Junction()
            self.csp = CumulativeSpectrum()

        with SignomialsEnabled():
            return [self.I <= self.jns.I,
                    self.V <= self.jns.V.sum(),
                    self.eta == self.V*self.I/total_irrad,
                    self.csp.lam == self.jns.lam_g,  # TODO: keyerror when this is subinplace?
                    [self.jns.phi[i] <= self.csp.phi[i] - self.jns.G[:i].sum()
                     for i in range(N_junctions)],
                    self.jns,
                    self.csp
                   ]

c = Cell(3)
m = Model(1/c.eta * c.jns.I.prod()**-0.001, [c])#, {c.jns.lam_g: [1100.0]})
#                          c["E_{gV}"][-1] == 0.66*units("V"),
#                          c["E_{gV}"][:-1] >= 1.44*units("V")])
try:
    print m.localsolve().table()
except ValueError:
    print m.solve().table()
```
